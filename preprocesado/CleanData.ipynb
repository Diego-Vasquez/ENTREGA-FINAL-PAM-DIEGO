```python
import numpy as np
import pandas as pd
import re
import emoji
import warnings 
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
string="\"#$%&'()*+-/:<=>@[\]^_`{|}~"
#!.,?¿
def remove_punctuation(text):
  text = text.replace('\n', ' ')
  text = emoji.get_emoji_regexp().sub(r' ', text)
  text_no_punc = re.sub(r'http\S+', ' ', text)
  text_no_punc = "".join([c for c in text if c not in string])
  return text_no_punc
```

### Cargar al dataset

```python
train = pd.read_csv('train_yelp_clean.csv')
val = pd.read_csv('val_yelp_clean.csv')
test = pd.read_csv('test_yelp_clean.csv')
```

```python
train.head(2)
```

### Cambiar el tipo de dato

```python
#Cambiar a int 
train['Score'] = train['Score'].astype(dtype=int)
val['Score'] = val['Score'].astype(dtype=int)
test['Score'] = test['Score'].astype(dtype=int)
```

### Eliminar puntuación

```python
train['Text'] = train['Text'].apply(lambda x: remove_punctuation(x))
val['Text'] = val['Text'].apply(lambda x: remove_punctuation(x))
```

```python
test['Text'] = test['Text'].apply(lambda x: remove_punctuation(x))
```

### Eliminar Nan

```python
train.dropna(subset=['Text'], inplace=True)
train.isnull().sum()
```

```python
val.dropna(subset=['Text'], inplace=True)
val.isnull().sum()
```

```python
test.dropna(subset=['Text'], inplace=True)
test.isnull().sum()
```

### Añadir Longitud del texto

```python
len_train = train.Text.str.split().apply(lambda x: len(x))
len_val = val.Text.str.split().apply(lambda x: len(x))
len_test = test.Text.str.split().apply(lambda x: len(x))
```

```python
train['longitud'] = len_train
val['longitud'] = len_val
test['longitud'] = len_test
```

### Quitar Longitud < 7

```python
train = train[train['longitud']>7]
val = val[val['longitud']>7]
test = test[test['longitud']>7]
```

### Guardar

```python
train.groupby('Score').count()
```

```python
train.to_csv('train_yelp_clean.csv', index=False)
val.to_csv('val_yelp_clean.csv', index=False)
test.to_csv('test_yelp_clean.csv', index=False)
```
