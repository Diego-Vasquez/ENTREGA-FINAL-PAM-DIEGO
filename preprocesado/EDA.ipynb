#Apply Sentiment

```python
tokenizer = AutoTokenizer.from_pretrained("finiteautomata/beto-sentiment-analysis")
model = AutoModelForSequenceClassification.from_pretrained("finiteautomata/beto-sentiment-analysis")
model.to('cuda:0')
print('ok')
```

```python
def clasificacion_review(model, tokenizer, review):
    inputs = tokenizer(review, truncation=True,
                                add_special_tokens=True,
                                return_token_type_ids=False,
                                padding="max_length",
                                return_attention_mask=True,
                                return_tensors="pt",
                                max_length=512).to('cuda:0')
    outputs = model(**inputs)
    logits = outputs.logits
    _, predict = torch.max(logits, dim=1)
    return predict.item()
```

```python
import csv
def inference_csv(model, tokenizer, name, df):
    fieldnames = ['Text', 'Score','Sentiment']
    with open(name, 'a', newline='') as csvfile:
        for i in range(len(df.Text)):
            prediction = clasificacion_review(model, tokenizer, df.Text[i])
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            print(f"{i}/{len(df.Text)}")
            writer.writerow({'Text':df.Text[i], 'Score':df.Score[i], 'Sentiment': prediction})
```
