```python
def homogenizar(df, random_state=42):
    una_estrella = sklearn.utils.shuffle(df[df['Score']==1], random_state=42).reset_index(drop=True)
    dos_estrella = sklearn.utils.shuffle(df[df['Score']==2], random_state=42).reset_index(drop=True)
    tres_estrella = sklearn.utils.shuffle(df[df['Score']==3], random_state=42).reset_index(drop=True)
    cuatro_estrella = sklearn.utils.shuffle(df[df['Score']==4], random_state=42).reset_index(drop=True)
    cinco_estrella = sklearn.utils.shuffle(df[df['Score']==5], random_state=42).reset_index(drop=True)

    min_len = min(len(una_estrella),len(dos_estrella), len(tres_estrella), 
                  len(cuatro_estrella), len(cinco_estrella))

    una_estrella = una_estrella[: min_len]
    dos_estrella = dos_estrella[: min_len]
    tres_estrella = tres_estrella[: min_len]
    cuatro_estrella = cuatro_estrella[: min_len]
    cinco_estrella = cinco_estrella[: min_len]
    
    new_data = pd.concat([una_estrella, dos_estrella, tres_estrella, 
                          cuatro_estrella, cinco_estrella])
    return sklearn.utils.shuffle(new_data, random_state=42).reset_index(drop=True)
```
